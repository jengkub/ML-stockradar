```python

# Raw Package
import numpy as np
import pandas as pd

#Data Source
import yfinance as yf

#Data viz
import plotly.graph_objs as go

#Interval required 1 minute
data = yf.download(tickers='UBER', period='1d', interval='1m')

#declare figure
fig = go.Figure()

#Candlestick
fig.add_trace(go.Candlestick(x=data.index,
                open=data['Open'],
                high=data['High'],
                low=data['Low'],
                close=data['Close'], name = 'market data'))

# Add titles
fig.update_layout(
    title='Uber live share price evolution',
    yaxis_title='Stock Price (USD per Shares)')

# X-Axes
fig.update_xaxes(
    rangeslider_visible=True,
    rangeselector=dict(
        buttons=list([
            dict(count=15, label="15m", step="minute", stepmode="backward"),
            dict(count=45, label="45m", step="minute", stepmode="backward"),
            dict(count=1, label="HTD", step="hour", stepmode="todate"),
            dict(count=3, label="3h", step="hour", stepmode="backward"),
            dict(step="all")
        ])
    )
)

#Show
fig.show()
```

```python
import csv
import datetime
#read csv
with open("test.csv",'r') as file:
    reader = csv.reader(file)
    for row in reader:
        if row[7] == 'NKE':
            last = row
#update
x = datetime.datetime.now()
count = 0
DayM = 0
DayMo365 = {'1':31,'2':28,'3':31,'4':30,'5':31,'6':30,'7':31,'8':31,'9':30,'10':31,'11':30,'12':31}
DayMo366 = {'1':31,'2':29,'3':31,'4':30,'5':31,'6':30,'7':31,'8':31,'9':30,'10':31,'11':30,'12':31}
LastDate = last[0].split()[0].split('-')
print(LastDate)
DiffMo = int(x.month) - int(LastDate[1])
DiffYe = int(x.year) - int(LastDate[0])
if DiffYe == 0:
    if DiffMo == 0:
        DiffDay = int(x.day) - int(LastDate[2])
        if DiffDay != 0:
            DiffDay = str(DiffDay) + 'd'
            data = yf.download(tickers='NKE', period=DiffDay, interval='1h')
    elif DiffMo != 0 :
        for u in range(DiffMo):
            DayM = DayM + DayMo365[str(int(LastDate[1])+count)]
            count += 1
        DiffDay = DayM - int(LastDate[2]) + int(x.day)
        DiffDay = str(DiffDay) + 'd'
        data = yf.download(tickers='NKE', period=DiffDay, interval='1h')
elif DiffYe != 0:
    dayly = 0
    dayn = 0
    for j in range(1,int(LastDate[1])):
        dayly = dayly + DayMo365[str(j)]
    for i in range(1,int(x.month)):
        dayn = dayn + DayMo365[str(i)]
    DiffDay = (365*DiffYe) - dayly + dayn - int(LastDate[2]) + int(x.day)
    DiffDay = str(DiffDay) + 'd'
    data = yf.download(tickers='NKE', period=DiffDay, interval='1h')

for i in data.index.day:
    if data.index.year[count] == int(LastDate[0]):
        if data.index.month[count] == int(LastDate[1]):
            if i == int(LastDate[2])+1:
                break
    count += 1

ok = last[0].split()
if ok[1] == '09:30:00':
    down = 6
elif ok[1] == '10:30:00':
    down = 5
elif ok[1] == '11:30:00':
    down = 4
elif ok[1] == '12:30:00':
    down = 3
elif ok[1] == '13:30:00':
    down = 2
elif ok[1] == '14:30:00':
    down = 1
elif ok[1] == '15:30:00':
    down = 0

count = count - down
data['ticker'] = 'NKE'
data = data.iloc[count:,:]
data
```

```python
#save 100 stock
from selenium import webdriver
from selenium.webdriver.common.by import By 

stock = []
 
op = webdriver.ChromeOptions()
op.add_argument('headless') 
driver = webdriver.Chrome(options=op) 
driver.get("https://www.set.or.th/th/market/index/set100/overview")

elements = driver.find_elements(By.XPATH, "//div[@class='symbol pt-1']")
for element in elements:
    stock.append(element.text)

driver.quit()

print(stock)
```

```python
SET100 = list()
for i in range(len(stock)-1):
    SET100.append(stock[i]+'.bk')
print(SET100)
```

```python
from bs4 import BeautifulSoup
import requests

stockg = []

for i in stock:
    url = requests.get("https://www.settrade.com/th/equities/quote/"+i+"/company-profile/profile")
    soup = BeautifulSoup(url.content, "html.parser")
    data = soup.find_all("div",{"class":"fs-14px text-middle-gray m-0"})
    stockg.append(data[5].text)
```

```python
stockg = [x.replace('\n', '').replace(' ','') for x in stockg if x is not None]

print(stockg)
len(stockg)
```
