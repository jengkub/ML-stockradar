```python
from selenium import webdriver
from selenium.webdriver.common.by import By 
import sqlite3
from datetime import datetime

# Create a SQL connection to our SQLite database
con = sqlite3.connect("stock.sqlite")
cur = con.cursor()

def link(alink, stock):
    all_link = []

    op = webdriver.ChromeOptions()
    op.add_argument('headless') 
    driver = webdriver.Chrome(options=op) 
    driver.get(alink)
    numlink = driver.find_elements(By.XPATH, '//*[@id="posts-container"]/li')
    if numlink == []:
        return False

    #get all news link in page
    for i in range(len(numlink)):
        link = driver.find_element(By.XPATH, '//*[@id="posts-container"]/li['+ str(i+1) +']/div/h2/a')
        all_link.append(link.get_attribute("href"))

    #scrap news
    for i in all_link:
        try:
            driver.get(i)
            get_url = driver.current_url
            title = driver.find_element(By.XPATH, "//*[@id='the-post']/header/div/h1")
            date = driver.find_element(By.XPATH, '//*[@id="single-post-meta"]/span')
            tag = driver.find_elements(By.XPATH, '//a[@rel="tag"]')

            date = date.text.split("/")
            date.reverse()
            if int(date[0]) < 2019:
                return False
            date = "-".join(date)
            date_format = "%Y-%m-%d"
            date_obj = datetime.strptime(date, date_format)

            
            for ticker in tag: 
                if ticker.text in stock:
                    print(get_url)
                    print(title.text)
                    print(date)   
                    print(ticker.text)
                    cur.execute("INSERT INTO stock_news VALUES (?, ?, ?, ?)",(date_obj, title.text, get_url, ticker.text,))
                    
            print('---------------------------------------------------')
            
        except:
            pass
    con.commit()

def nextpage(alink, ticker):
    num = 1
    op = webdriver.ChromeOptions()
    op.add_argument('headless') 
    # op.add_argument('headless') 
    driver = webdriver.Chrome(options=op) 
    #next page
    driver.get(alink)

    while True:
        num += 1
        run = link('https://www.kaohoon.com/tag/'+ticker+'/page/'+str(num), stock)
        if run == False:
            print('stop')
            break






# driver.quit()
```
