```python
#save 100 stock
from selenium import webdriver
from selenium.webdriver.common.by import By 

stock = []
 
op = webdriver.ChromeOptions()
op.add_argument('headless') 
driver = webdriver.Chrome(options=op) 
driver.get("https://www.set.or.th/th/market/index/set100/overview")

elements = driver.find_elements(By.XPATH, "//div[@class='symbol pt-1']")
for element in elements:
    stock.append(element.text)

driver.quit()

print(stock)
```

```python
SET100 = list()
for i in range(len(stock)-1):
    SET100.append(stock[i]+'.BK')
print(SET100)
```

```python
from bs4 import BeautifulSoup
import requests

stockgroup = []
substockgroup = []

for i in stock:
    url = requests.get("https://www.settrade.com/th/equities/quote/"+i+"/company-profile/profile")
    soup = BeautifulSoup(url.content, "html.parser")
    data = soup.find_all("div",{"class":"fs-14px text-middle-gray m-0"})
    stockgroup.append(data[5].text)
    substockgroup.append(data[6].text)
```

```python
stockgroup = [x.replace('\n', '').replace(' ','') for x in stockgroup if x is not None]
substockgroup = [x.replace('\n', '').replace(' ','') for x in substockgroup if x is not None]

print(stockgroup)
print(substockgroup)
len(stockgroup)
```
