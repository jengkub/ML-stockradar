```python
#save 100 stock
from selenium import webdriver
from selenium.webdriver.common.by import By 

stock = []
 
op = webdriver.ChromeOptions()
op.add_argument('headless') 
driver = webdriver.Chrome(options=op) 
driver.get("https://www.set.or.th/th/market/index/set100/overview")

elements = driver.find_elements(By.XPATH, "//div[@class='symbol pt-1']")
for element in elements:
    stock.append(element.text)

driver.quit()

print(stock)
```

```python
SET100 = list()
for i in range(len(stock)):
    SET100.append(stock[i]+'.BK')
print(SET100)
```

```python
from bs4 import BeautifulSoup
import requests

stockgroup = []
substockgroup = []

for i in stock:
    url = requests.get("https://www.settrade.com/th/equities/quote/"+i+"/company-profile/profile")
    soup = BeautifulSoup(url.content, "html.parser")
    data = soup.find_all("div",{"class":"fs-14px text-middle-gray m-0"})
    stockgroup.append(data[5].text)
    substockgroup.append(data[6].text)
```

```python
stockgroup = [x.replace('\n', '').replace(' ','') for x in stockgroup if x is not None]
substockgroup = [x.replace('\n', '').replace(' ','') for x in substockgroup if x is not None]

print(stockgroup)
print(substockgroup)
len(stockgroup)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By 

head= []
finacetype = []
finace = []
string = []
 
op = webdriver.ChromeOptions()
op.add_argument('headless') 
driver = webdriver.Chrome(options=op) 
driver.get("https://www.settrade.com/th/equities/quote/ACE/financial-statement/five-years")

elements = driver.find_elements(By.XPATH, "//table/thead/tr/th/div")
for element in elements:
    head.append(element.text)

elements = driver.find_elements(By.XPATH, "//table/tbody/tr/th")
for element in elements:
    finacetype.append(element.text)

elements = driver.find_elements(By.XPATH, "//table/tbody/tr/td[@class='column-sticky']")
for element in elements:
    finace.append(element.text)

elements = driver.find_elements(By.XPATH, "//table/tbody/tr/td/div/div")
for element in elements:
    string.append(element.text)

driver.quit()
```

```python
import sqlite3
import datetime
conn = sqlite3.connect("stock.sqlite")
cur = conn.cursor()
ticker = 'AOT.BK'
query = "select * from stock_table where `Ticker` == '%s'" % ticker
r_df = pd.read_sql(query,conn)
last = r_df.tail(1).Datetime.to_string().split()
LastDate = last[1].split()[0].split('-')
print(LastDate)
x = datetime.datetime.now()
count = 0
DayM = 0
down = 0
DayMo365 = {'1':31,'2':28,'3':31,'4':30,'5':31,'6':30,'7':31,'8':31,'9':30,'10':31,'11':30,'12':31}
DiffMo = int(x.month) - int(LastDate[1])
DiffYe = int(x.year) - int(LastDate[0])
if DiffYe == 0:
    if DiffMo == 0:
        DiffDay = int(x.day) - int(LastDate[2])
        if DiffDay != 0:
            pass
    elif DiffMo != 0 :
        for u in range(DiffMo):
            DayM = DayM + DayMo365[str(int(LastDate[1])+count)]
            count += 1
        DiffDay = DayM - int(LastDate[2]) + int(x.day)
elif DiffYe != 0:
    dayly = 0
    dayn = 0
    for j in range(1,int(LastDate[1])):
        dayly = dayly + DayMo365[str(j)]
    for i in range(1,int(x.month)):
        dayn = dayn + DayMo365[str(i)]
    DiffDay = (365*DiffYe) - dayly + dayn - int(LastDate[2]) + int(x.day)
    
DiffDay = str(DiffDay) + 'd'
data = yf.download(tickers=ticker, period=DiffDay, interval='1h')
print(DiffDay)
for i in data.index.day:
    if data.index.year[count] == int(LastDate[0]):
        if data.index.month[count] == int(LastDate[1]):
            if i == int(LastDate[2])+1:
                break
    count += 1

ok = r_df.tail(1).Datetime.to_string().split()[2]
if ok == '10:00:00':
    down = 5
elif ok == '11:00:00':
    down = 4
elif ok == '12:00:00':
    down = 3
elif ok == '14:00:00':
    down = 2
elif ok == '15:00:00':
    down = 1
elif ok == '16:00:00':
    down = 0

count = count - down
data['ticker'] = ticker
data = data.iloc[count:,:]
print(data)
data.to_sql('stock_table',con=conn,if_exists='append',index=True)
cur.close()
```

```python
import sqlite3

# Create a SQL connection to our SQLite database
con = sqlite3.connect("stock.sqlite")

cur = con.cursor()
cur.execute("CREATE TABLE stock_table_mo (Datetime DATETIME, Open real, High real, Low real, Close real, Adj Close real, Volume integer, Ticker VARCHAR(255), FOREIGN KEY(Ticker) REFERENCES stock_info(Ticker));")
con.commit()
```

Insert database

```python
con = sqlite3.connect("stock.sqlite")

df_list = list()
cur = con.cursor()
for ticker in SET100:
    data = yf.download(tickers=ticker, period='max', interval='1mo',progress=False)
    for d in range(len(data.index)):
        date_format = '%Y-%m-%d %H:%M:%S'
        date_obj = datetime.strptime(str(data.index[d]), date_format)
        cur.execute("INSERT INTO stock_table_mo (Datetime, Open, High, Low, Close, Adj, Volume, Ticker) VALUES (? ,? ,? ,? ,? ,? ,? ,? )",(date_obj, data.Open[d], data.High[d], data.Low[d], data.Close[d], data['Adj Close'][d], int(data.Volume[d]), ticker,))
con.commit()
```
