```python

# Raw Package
import numpy as np
import pandas as pd

#Data Source
import yfinance as yf

#Data viz
import plotly.graph_objs as go
import sqlite3

conn = sqlite3.connect("test.sqlite")
cur = conn.cursor()
ticker = 'AOT.BK'
query = "select * from stock_table where `ticker` == '%s'" % ticker
r_df = pd.read_sql(query,conn)
#declare figure
fig = go.Figure()

#Candlestick
fig.add_trace(go.Candlestick(x=r_df.Datetime,
                open=r_df.Open,
                high=r_df.High,
                low=r_df.Low,
                close=r_df.Close, name = 'market data'))

# Add titles
fig.update_layout(
    title='AOT live share price evolution',
    yaxis_title='Stock Price (THB per Shares)')

# X-Axes
fig.update_xaxes(
    rangeslider_visible=True,
    rangeselector=dict(
        buttons=list([
            dict(count=15, label="15m", step="minute", stepmode="backward"),
            dict(count=45, label="45m", step="minute", stepmode="backward"),
            dict(count=1, label="HTD", step="hour", stepmode="todate"),
            dict(count=3, label="3h", step="hour", stepmode="backward"),
            dict(step="all")
        ])
    )
)

#Show
fig.show()
```

```python
#save 100 stock
from selenium import webdriver
from selenium.webdriver.common.by import By 

stock = []
 
op = webdriver.ChromeOptions()
op.add_argument('headless') 
driver = webdriver.Chrome(options=op) 
driver.get("https://www.set.or.th/th/market/index/set100/overview")

elements = driver.find_elements(By.XPATH, "//div[@class='symbol pt-1']")
for element in elements:
    stock.append(element.text)

driver.quit()

print(stock)
```

```python
SET100 = list()
for i in range(len(stock)-1):
    SET100.append(stock[i]+'.BK')
print(SET100)
```

```python
from bs4 import BeautifulSoup
import requests

stockgroup = []
substockgroup = []

for i in stock:
    url = requests.get("https://www.settrade.com/th/equities/quote/"+i+"/company-profile/profile")
    soup = BeautifulSoup(url.content, "html.parser")
    data = soup.find_all("div",{"class":"fs-14px text-middle-gray m-0"})
    stockgroup.append(data[5].text)
    substockgroup.append(data[6].text)
```

```python
stockgroup = [x.replace('\n', '').replace(' ','') for x in stockgroup if x is not None]
substockgroup = [x.replace('\n', '').replace(' ','') for x in substockgroup if x is not None]

print(stockgroup)
print(substockgroup)
len(stockgroup)
```

แบ่งหมวดหมู่กลุ่มอุตสาหกรรม

```python
con = sqlite3.connect("stockinfo.db")

cur = con.cursor()
for i, j, k in zip(stock, stockgroup, substockgroup):
    cur.execute("INSERT INTO Info VALUES (? ,? ,?)",(i, j, k,))
con.commit()
```
